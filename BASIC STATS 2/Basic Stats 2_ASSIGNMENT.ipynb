```python
#Import libraries
import numpy as np
import scipy.stats as stats
```

```python
# Sample data
data=np.array([1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29])
```

##### a. Build 99% Confidence Interval Using Sample Standard Deviation

```python
#Given data
n = len(data)
print("Sample Size of data: ",n)
df=n-1
print("Degree of freedom:",df)
# CI=0.99 , alpha=0.01
```

```python
mean = np.mean(data)
print("Sample Mean: ",mean)
```

```python
s=sample_std = np.std(data, ddof=1)
print("Sample standard deviation: ",s)
```

#### Build 99% Confidence Interval Using Sample Standard Deviation

```python
t_critical_lower=stats.t.ppf(0.005,14)
t_critical_upper=stats.t.ppf(0.995,14)
print(t_critical_lower,t_critical_upper)
```

```python
# Margin of error
Margin_lower=t_critical_lower*(s/np.sqrt(n))
Margin_upper=t_critical_upper*(s/np.sqrt(n))
print(Margin_lower,Margin_upper)
```

```python
confidence_interval = (mean + Margin_lower, mean + Margin_upper)
confidence_interval
```

- When we do not know the population standard deviation, we use the sample standard deviation instead.
- Because the sample size is small, we use the t-distribution to make the calculation more accurate.
- Using this method, the 99% confidence interval gives a range of values where the true average durability is likely to be found.
- This range also shows the uncertainty because the data comes from a small sample.

#### Build 99% Confidence Interval Using Known Population Standard Deviation

```python
# Given
pop_std=0.2
```

```python
# Z critical value
z_critical_lower=stats.norm.ppf(0.005)
z_critical_upper=stats.norm.ppf(0.995)
print(z_critical_lower,z_critical_upper)
```

```python
# Margin of error
margin_z_lower = z_critical_lower * (pop_std / np.sqrt(n))
margin_z_upper = z_critical_upper * (pop_std / np.sqrt(n))
print(margin_z_lower,margin_z_upper)
```

```python
# Confidence interval
lower_z = mean + margin_z_lower
upper_z = mean + margin_z_upper
print(lower_z,upper_z)
```

- If the population standard deviation is known, the z-distribution is used.
- In this case, the confidence interval is slightly narrower because the variability in the estimate is lower.
- Both approaches give a statistical range around the sample mean, allowing the manufacturer to make informed quality control decisions about the print-head durability.

#### Conclusion:

1. The resulting confidence intervals provide a range in which the true mean durability of the print-heads is likely to lie.
2. Using the t-distribution accounts for uncertainty from the small sample size and unknown population standard deviation, while using the z-distribution with a known standard deviation provides a slightly narrower, more precise estimate.
3. Both approaches help the manufacturer make informed decisions regarding product reliability and quality.
