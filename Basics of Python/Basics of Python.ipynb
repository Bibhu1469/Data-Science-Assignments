### Exercise 1: Prime Numbers

```python
a=int(input('Enter a number: '))
if a>1:
    for i in range(2,a):
        if a%i==0:
            print(a,'is not a prime number')
            break
    else:
            print(a,'is a prime number')
else:
    print(a,'is not a prime number')
        
```

### Exercise 3: Squares of Even/Odd Numbers

```python
print("Squares of all even numbers from 100 to 200:")
for num in range(100, 201):
    if num % 2 == 0:
        square = num * num
        print('The square of',num,'is',square)
```

```python
print("Squares of all odd numbers from 100 to 200:")
for num in range(100, 201):
    if num % 2 != 0:
        square = num * num
        print('The square of',num,'is',square)
```

### Exercise 4: Word counter

```python
text =input('Enter your text.')

word_count = {}

corrected_text = text.lower().replace('.', '')

words = corrected_text.split()

for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

for word, count in word_count.items():
    print(f"'{word}': {count}")
```

### Exercise 5: Check for Palindrome

```python
text=input('Enter word')
#revised_text=text.lower().replace(',','','!').replace(' ','','')
revised_text = text.lower().replace(',', '').replace(' ', '').replace('!', '').replace('.', '')
print(revised_text)
reversed_text = revised_text[::-1]
if revised_text == reversed_text :
    print('Yes')
else:
    print('No')
```
